```python
# Get one-hot encoding for categorical variables
data_encoded = pd.get_dummies(df_new, columns=['fuel', 'seller_type', 'transmission', 'owner'], drop_first=True)

data_encoded.head()
```

```python
data_encoded.info()
```

```python
encoded_df = data_encoded
encoded_df.info()
```

```python
encoded_df.columns
```

```python
# Extracting features for training and testing model
X = encoded_df[['km_driven', 'mileage_kmpl', 'engine_cc', 'max_power_bhp', 'seats', 'age', 'fuel_Diesel', 'fuel_LPG',
       'fuel_Petrol', 'seller_type_Individual', 'seller_type_Trustmark Dealer',
       'transmission_Manual', 'owner_Fourth & Above Owner',
       'owner_Second Owner', 'owner_Test Drive Car', 'owner_Third Owner']]
Y = encoded_df['selling_price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()
reg_model.fit(x_train, y_train)
reg_y_pred = reg_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, accuracy_score

print('Linear Regression Evaluation')
print()
print('MAE: ', mean_absolute_error(y_test, reg_y_pred))
print('MSE: ', mean_squared_error(y_test, reg_y_pred))
print('r2 score: ', r2_score(y_test, reg_y_pred))
```

```python
from sklearn.linear_model import Ridge, Lasso
#ridge regression
ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)

ridge_y_pred = ridge.predict(x_test)
print('Ridge Regression Evaluation')
print()
print('mae: ', mean_absolute_error(y_test, ridge_y_pred))
print('mse: ', mean_squared_error(y_test, ridge_y_pred))
print('r2 score: ', r2_score(y_test, ridge_y_pred))
```

```python
# Perform Lasso regression
lasso = Lasso(alpha=939)
lasso.fit(x_train, y_train)

lasso_y_pred = lasso.predict(x_test)
print('Lasso Regression Evaluation')
print()
print('mae: ', mean_absolute_error(y_test, lasso_y_pred))
print('mse: ', mean_squared_error(y_test, lasso_y_pred))
print('r2 score: ', r2_score(y_test, lasso_y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_scores=[]
estimators=[10,50,100,200,500,1000]
for i in estimators:
  model=RandomForestRegressor(n_estimators=i)
  model.fit(x_train,y_train)
  score=model.score(x_test,y_test)
  rf_scores.append(score)
  y_pred=model.predict(x_test)
  mae=mean_absolute_error(y_test,y_pred)
  print(f"{i} Estimators: " + "Mean Absolute Error: ", mae)
  y_avg=y_test.mean()
  mape=(mae/y_avg)*100
  print(f"{i} Estimators: " + "Mean Absolute Percentage Error: %.2f%%" %mape)
  mse=mean_absolute_error(y_test,y_pred)
  print(f"{i} Estimators: " + "Mean Absolute Error: ", mse)
  r2=r2_score(y_test,y_pred)
  print(f"{i} Estimators: " + "r2 score: ", r2)
```

```python
rf_scores
```

```python
from matplotlib import rcParams
from matplotlib.cm import rainbow
colors = rainbow(np.linspace(0, 1, len(estimators)))
plt.bar([i for i in range(len(estimators))], rf_scores, color = colors, width = 0.8)
for i in range(len(estimators)):
    plt.text(i, rf_scores[i], round(rf_scores[i], 2))
plt.xticks(ticks = [i for i in range(len(estimators))], labels = [str(estimator) for estimator in estimators])
plt.xlabel('Number of estimators')
plt.ylabel('Scores')
plt.title('Random Forest Classifier scores for different number of estimators')
```

**Conclusion:** Car Selling Price is predicted more accurately by Random Forest Regressor with 94% accuracy
